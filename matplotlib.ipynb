```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def on_press(event):
    if event.inaxes is None: return
    for line in event.inaxes.lines:
        if event.key=='t':
            visible = line.get_visible()
            line.set_visible(not visible)
    event.inaxes.figure.canvas.draw()

fig, ax = plt.subplots(1)

fig.canvas.mpl_connect('key_press_event', on_press)

ax.plot(np.random.rand(2, 20))

plt.show()
```

### Plotting without scripting layer

```python
# First let's set the backend without using mpl.use() from the scripting layer
from matplotlib.backends.backend_agg import FigureCanvasAgg
from matplotlib.figure import Figure
import numpy as np

# create a new figure
fig = Figure()

# associate fig with the backend
canvas = FigureCanvasAgg(fig)

# generate 10000 normaly distributed sample points 
x = np.random.randn(10000)

# add a subplot to the fig
ax = fig.add_subplot(1,1,1)

# plot the histogram for x array
ax.hist(x, 100)

# save the figure to histogram.png
ax.set_title("My First Histogram")   # set the title of the plot as "My First Histogram"
fig.savefig('histogram.png')   
```

## Basic Plots using Matplotlib

```python
iris = pd.read_csv("iris.csv")
iris.head()
```

## 1. Scatter Plots

```python
plt.figure()
plt.scatter(iris['petal_width'], iris['petal_length'],s=50, c='red')
```

```python
# changing the size and color of the points based on species
plt.figure()
sp1 = iris[iris['species']=='setosa']
sp2 = iris[iris['species']=='virginica']
sp3 = iris[iris['species']=='versicolor']
plt.scatter(sp1['petal_width'], sp1['petal_length'], s= 25, c='red', label = 'setosa')
plt.scatter(sp2['petal_width'], sp2['petal_length'], s= 25, c='blue', label = 'virginica')
plt.scatter(sp3['petal_width'], sp3['petal_length'], s= 25, c='green', label = 'versicolor')
```

```python
plt.title('Petal Width vs Petal Length (Iris Dataset)')
plt.xlabel('Petal Width')
plt.ylabel('Petal Length')
plt.legend(['S','Vi','Ver'])
```

## Subplots

```python
plt.figure()
# subplot with 1 row, 2 columns, and current axis is 1st subplot axes
plt.subplot(1, 2, 1)
plt.scatter(iris['petal_width'], iris['petal_length'])
```

```python
plt.subplot(1, 2, 2)
plt.hist(iris['petal_width'],bins=30);
```

```python
plt.figure()
ax1 = plt.subplot(1, 2, 1)
plt.scatter(iris['petal_width'], iris['petal_length'])
# pass sharey=ax1 to ensure the two subplots share the same y axis
ax2 = plt.subplot(1, 2, 2, sharey=ax1)
plt.hist(iris['petal_width'],bins=30);
```

```python
# create 2x2 grid of axis subplots
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True)
axs = [ax1,ax2,ax3,ax4]

# draw n = 10, 100, 1000, and 10000 samples from the normal distribution and plot corresponding histograms
for n in range(0,len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    axs[n].hist(sample)
    axs[n].set_title('n={}'.format(sample_size))
```

```python
import matplotlib.gridspec as gridspec
```

```python
# use gridspec to partition the figure into subplots
plt.figure()
gspec = gridspec.GridSpec(3, 3)

top_histogram = plt.subplot(gspec[0, 1:])
side_histogram = plt.subplot(gspec[1:, 0])
lower_right = plt.subplot(gspec[1:, 1:])
```

```python
Y = np.random.normal(loc=0.0, scale=1.0, size=10000)
X = np.random.random(size=10000)
lower_right.scatter(X, Y)
top_histogram.hist(X, bins=100)
side_histogram.clear()
side_histogram.hist(Y, bins=100, orientation='horizontal')
side_histogram.invert_xaxis()
```
