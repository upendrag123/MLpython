## 2. Kmeans Clustering

### Kmeans++
The exact algorithm is as follows:

1.Choose one center uniformly at random from among the data points. <br>
2.For each data point x, compute D(x), the distance between x and the nearest center that has already been chosen.<br>
3.Choose one new data point at random as a new center, using a weighted probability distribution where a point x is chosen with probability proportional to D(x)^2. <br>
4.Repeat Steps 2 and 3 until k centers have been chosen. <br>
5.Now that the initial centers have been chosen, proceed using standard k-means clustering.<br>

```python
km = kmeans(n_clusters=10,random_state=42)
```

```python
km.fit(movies.iloc[:,1:20])
```

```python
 km.labels_
```

```python
movies['kmeansclust'] = km.labels_
```

```python
movies.head()
```

```python
km.n_iter_
```

```python
# Sum of distances of samples to their closest cluster center.
km.inertia_
```

```python
pd.Series(km.labels_).value_counts().sort_index()
```

```python
km.cluster_centers_
```

```python
pd.DataFrame(km.cluster_centers_.round(2),columns=movies.columns[1:20],index=['Act-Adv-Scifi-Thr','Drama','Comedy','Horror','Rom-Com','Cri-Mys-Thr','Documentary','Rom-Drama','Adv-Anim-Child','Com-Drama'])
```

```python
movies.groupby('kmeansclust').mean()
```

#### Scree Plot for finding number of clusters

```python
withinss =[]
for i in range(2,21):
    km=kmeans(n_clusters=i, n_init=10,random_state=42)
    withinss.append(km.fit(movies.iloc[:,1:20]).inertia_)
```

```python
plt.plot(range(2,21),withinss,'-o')
plt.xlabel("Num of Clusters")
plt.ylabel("Total Within SS")
plt.xlim(1,21)
```
