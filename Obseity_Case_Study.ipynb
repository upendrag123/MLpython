6. Add a variable to our USDA dataframe that takes a value 1 if the food has higher sodium than average,and 0 if the food has lower sodium than average. 
Note: Similarly add the variables HighProtein, HighCarbs, HighFat, to our data frame.

```python
USDA['Sodium_Hi_Lo_Flg'] = [int(i) for i in (USDA['Sodium']> 322)]
USDA['HighProtein'] = [int(i) for i in (USDA['Protein']> 10.9)]
USDA['HighCarbs'] = [int(i) for i in (USDA['Carbohydrate']> 27.63)]
USDA['HighFat'] = [int(i) for i in (USDA['TotalFat']> 16.81)]
USDA
```

7. How many foods have higher Sodium level than average?


```python
USDA['Sodium_Hi_Lo_Flg'].sum()
```

8. How many foods have both high Sodium and high Fat level than average?

```python
print('Foods High in Sodium and Fat=',USDA[USDA['Sodium_Hi_Lo_Flg'] + USDA['HighFat']>1]['HighFat'].sum())
```
